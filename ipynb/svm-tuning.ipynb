```python
from classifier import SVM
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import MultipleLocator
from tqdm import tqdm
```

```python
accu = []
f1 = []
c_range = np.arange(0.5, 10, 0.5)
for i in tqdm(c_range):
    clf = SVM(i)
    clf.train()
    clf.score()
    accu.append(clf.accu)
    f1.append(clf.f1)
```

```python
plt.figure(figsize=(8, 4), dpi=80)
plt.plot(c_range, accu, color="red", label="accuracy-n_neighbors curve")
plt.plot(c_range, f1, color="blue", label="f1 score-n_neighbors curve")
plt.xlabel('n_neighbors')
plt.ylabel('accuracy')

x_major_locator = MultipleLocator(1)
ax = plt.gca()
ax.xaxis.set_major_locator(x_major_locator)
ax.set_title('relationship between accuracy and number of c', fontsize=14)

plt.grid()
plt.legend(loc='best')
plt.show()
```

```python
print('best accuracy:{}\nbest f1-score:{}'.format(accu.index(max(accu)), f1.index(max(f1))))
```
