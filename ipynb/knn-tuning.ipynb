```python
from classifier import KNN
import matplotlib.pyplot as plt
from matplotlib.pyplot import MultipleLocator
from tqdm import trange
```

knn调试参数

```python
accu = []
f1 = []
k_range = range(1, 30)
for i in trange(1, 30):
    clf = KNN(i + 1)
    clf.train()
    clf.score()
    accu.append(clf.accu)
    f1.append(clf.f1)
```

输出结果

```python
plt.figure(figsize=(8, 4), dpi=80)
plt.plot(k_range, accu, color="red", label="accuracy-n_neighbors curve")
plt.plot(k_range, f1, color="blue", label="f1 score-n_neighbors curve")
plt.xlabel('n_neighbors')
plt.ylabel('evaluation value')

x_major_locator = MultipleLocator(1)
ax = plt.gca()
ax.xaxis.set_major_locator(x_major_locator)
ax.set_title('relationship between performance and number of n_neighbors', fontsize=14)

plt.grid()
plt.legend(loc='best')
plt.show()
```

```python
print('best accuracy at:{}\nbest f1-score at:{}'.format(accu.index(max(accu)), f1.index(max(f1))))
```
